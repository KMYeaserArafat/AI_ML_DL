You are given height_weight.csv file which contains heights and weights of 1000 people. Dataset is taken from here, https://www.kaggle.com/mustafaali96/weight-height

You need to do this,

1) Load this csv in pandas dataframe and first plot histograms for height and weight parameters

2) Using IQR detect weight outliers and print them

3) Using IQR, detect height outliers and print them

```python
import pandas as pd

df = pd.read_csv('weight-height.csv')
df.head()
```

```python
# Height, 

import matplotlib.pyplot as plt

plt.hist(df['Height'], bins=20, rwidth=0.3)
plt.xlabel("Height")
plt.ylabel("Count")
plt.show()
```

```python
# Weights,
import matplotlib.pyplot as plt

plt.hist(df['Weight'], bins=20, rwidth=0.3)
plt.xlabel("Weight")
plt.ylabel("Count")
plt.show()
```

```python
Q1 = df['Weight'].quantile(0.25)
Q3 = df['Height'].quantile(0.75)


IQR = Q3-Q1
IQR
```

```python
lower_bound = Q1 - 1.5*IQR
Upper_bound = Q3 + 1.5*IQR
```

Weight Outlier

```python
weight_outlier = df[(df['Weight']>Upper_bound) | (df['Weight']<lower_bound)]
weight_outlier.shape
```

```python
df1 = df[(df['Weight']<Upper_bound) & (df['Weight']>lower_bound)]
df1.shape
```

Height Outlier

```python
height_outlier = df[(df['Height']>Upper_bound) | df['Height']<lower_bound]
height_outlier.shape
```
