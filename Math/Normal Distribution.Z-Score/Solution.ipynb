```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bhp.csv")
df.head()
```

```python
plt.hist(df['price_per_sqft'],bins=20, rwidth=0.8)
plt.xlabel("Price per Square")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['price_per_sqft'],bins=20, rwidth=0.8)
plt.xlabel("Price per Square")
plt.ylabel("Count")
plt.yscale('log')
plt.show()
```

```python
# 1) 
lowerBound = df.price_per_sqft.quantile(0.01)
uppderBound = df.price_per_sqft.quantile(0.999)

bound_df = df[(df.price_per_sqft>lowerBound) & (df.price_per_sqft<uppderBound)]
bound_df
```

```python
removal = df.shape[0] - bound_df.shape[0]
print(f"Total Removal : {removal} ")
```

```python
plt.hist(bound_df.price_per_sqft,bins=20, rwidth=0.8)
plt.xlabel("Price per Square")
plt.ylabel("Count")
plt.show()
```
