**2. What is the average age (*age* feature) of women?**

**3. What is the percentage of German citizens (*native-country* feature)?**

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
earn_50K_more = data[(data['salary'] == ">50K")]
earn_50K_less = data[(data['salary'] == '<=50K')]


def mean_std(title,data,feature):
    mean_data = data[feature].mean()
    std_data = data[feature].std()
    print(f"{title} mean :  {mean_data}")
    print(f"{title} Standard deviation : {std_data} \n")


mean_std("50K More Salary", earn_50K_more, 'age')
mean_std("50K Less Salary", earn_50K_less, 'age')
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
# Step 1: Group by 'race' and 'sex', then describe the 'age' column
age_statistics = data.groupby(['race', 'sex'])['age'].describe()

# Display age statistics
print(age_statistics)

# Step 2: Find the maximum age of men of the 'Amer-Indian-Eskimo' race
max_age_amerindian_men = data[(data['race'] == 'Amer-Indian-Eskimo') & (data['sex'] == 'Male')]['age'].max()

# Output the maximum age
print(f"Maximum age of Amer-Indian-Eskimo men: {max_age_amerindian_men}")
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
# Step 1: Filter data for men only
men_data = data[data['sex']== 'Male']

# Step 2: Create a new column 'marital-status-group' to classify men as 'Married' or 'Single'
men_data['marital-status-group'] = men_data['marital-status'].apply(lambda x: 'Married' if x.startswith('Married') else 'Single')

# Step 3: Filter men who earn more than 50K
men_earn_more_50K = men_data[(men_data['salary'] == ">50K")]

# Step 4: Calculate the number of married and single men who earn more than 50K
married_man_earn_more_50K = men_earn_more_50K[(men_earn_more_50K['marital-status-group'] == 'Married')].shape[0]
single_men_earn_more_50K = men_earn_more_50K[(men_earn_more_50K['marital-status-group'] == 'Single')].shape[0]

# Step 5: Calculate the total number of married and single men
married_total_man = men_data[(men_data['marital-status-group']=='Married')].shape[0]
single_total_man = men_data[(men_data['marital-status-group']=='Single')].shape[0]


# Step 6: Calculate the proportion of married and single men who earn more than 50K
married_proportion = round((married_man_earn_more_50K/married_total_man)*100)
single_proportion = round((single_men_earn_more_50K/single_total_man)*100)


print(f"Married Proportion : {married_proportion}%")
print(f"Single Proportion : {single_proportion}%")

import matplotlib.pyplot as plt
plt.pie([married_proportion, single_proportion],labels=["Married", "Single"])
plt.legend([f" 50K+ Salary holder {married_proportion}%", f"50K+ Salary holder {single_proportion}%"])
plt.show()
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
maximum_work_per_week = data['hours-per-week'].max()
maximum_person_50K_Salary = data[(data['hours-per-week']== maximum_work_per_week) & (data['salary'] == ">50K")]

print(f"50K+ Salary holder Whose work per Week {maximum_work_per_week} Hours, {maximum_person_50K_Salary}")

import matplotlib.pyplot as plt

plt.pie([maximum_person_50K_Salary.shape[0], data.shape[0]], labels=['Hark Worker', 'Normal Worker'])
plt.legend([f"Hard Worker Employee : {maximum_person_50K_Salary.shape[0]}" , f"Normal Worker Employee {data.shape[0]}"])
plt.show()
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python

# Step 1: Filter the dataset based on salary
low_salary_data = data[data['salary'] == "<=50K"]
high_salary_data = data[data['salary'] == ">50K"]

# Step 2: Group by 'native-country' and calculate the average 'hours-per-week'
average_hours_low_salary = low_salary_data.groupby('native-country')['hours-per-week'].mean().reset_index()
average_hours_low_salary.columns = ['native-country', 'average_hours_low_salary']

average_hours_high_salary = high_salary_data.groupby('native-country')['hours-per-week'].mean().reset_index()
average_hours_high_salary.columns = ['native-country', 'average_hours_high_salary']

# Step 3: Merge the results on 'native-country'
average_hours_combined = pd.merge(average_hours_low_salary, average_hours_high_salary, on='native-country', how='outer')

# Display the results
print(average_hours_combined)
```
