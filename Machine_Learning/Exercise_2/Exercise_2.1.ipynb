```python
import pandas as pd
from sklearn import linear_model
```

```python
data = pd.read_csv("homeprices.csv")
data.head()
```

```python
# replace NaN by medium, 
median_value = data['bedrooms'].median()

data['bedrooms'] =  data['bedrooms'].fillna(median_value)
data.head()
```

```python
reg = linear_model.LinearRegression()
x = data[['area', 'bedrooms', 'age']]
y = data['price']
reg.fit(x,y)
```

```python
# Find Coefficent, 
m = reg.coef_
print(m)

# Find intercept, 
b = reg.intercept_
print(b)
```

### Without using predict method, 
### Find Manually

```python
def PredictionPrice(area,bedRoom,age): 
    price = ((m[0]*area) + (m[1]*bedRoom) + (m[2]*age)) + b
    print(f"{area} sqr ft area, {bedRoom} bedrooms, {age} year old Predicted Price : {price}")


# 3000 sqr ft area, 3 bedrooms, 40 year old
PredictionPrice(3000,3,40)

# 2500 sqr ft area, 4 bedrooms, 5 year old
PredictionPrice(2500,3,5)
```

### using predict Method, 

```python
# 3000 sqr ft area, 3 bedrooms, 40 year old
result1 = reg.predict([[3000,3,40]])
print(f"3000 sqr ft area, 3 bedrooms, 40 year old Predicted Price : {result1}")

## 2500 sqr ft area, 4 bedrooms, 5 year old
result2 = reg.predict([[2500,4,5]])
print(f"2500 sqr ft area, 4 bedrooms, 5 year old Predicted Price : {result2}")
```
